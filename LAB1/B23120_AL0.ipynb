Question 3: Array adressing and slicing. As the dimensions of the array increase it becomes very important how one indexes those arrays, slicing becomes important when one wants to get some desired subpart of a whole array. This Question has two parts.

3.1. Print the sum of left diagonal and right diagonal of a square matrix. 

```python
mat= np.array([[1,3,4,5,2],[1,5,2,4,3],[5,2,3,4,1,],[1,4,2,6,9],[4,5,2,1,7]])

def left_diagonal_sum(mat:np.ndarray)->float:
    #write your logic here
    sum=0
    r,c = mat.shape  # returns a tuple containing number of rows and columns of the 2D matrix
    for i in range(r):
        sum += mat[i][i]  # add first element of first row, second element of 2nd row and so on
    return sum


def right_diagonal_sum(mat:np.ndarray)->float:
    #write your logic here
    sum=0
    r,c = mat.shape
    for i in range(r):
        sum += mat[i][r-i-1]  # adds last element of first row, second last element of second row and so on
    return sum

print(f'Left Diagonal Sum of {mat=} is {left_diagonal_sum(mat)}')
print(f'Left Diagonal Sum of {mat=} is {right_diagonal_sum(mat)}')
```

3.2. From the matrix of previous part, print the 3x4 sub-matrix from 1st row 2nd column to 3rd row 5th column. Make use of the array slicing operations preferably , use of for loops is discouraged. 

```python
def submatrix_3x4(mat: np.ndarray)-> np.ndarray:
    #write your logic here 
    return mat[:3, 1:]
    # :3 specfies the rows to be included: 0th to 2nd while 1: specifies to take all columns except the first (col 0)

print(f'The desired submatrix of \n {mat} is \n {submatrix_3x4(mat)}')
```

Question 4:  array operations in np.array and np math functions. Plot the function: $$ f(x) = {sin^7(x) + cos^5(x) \over e^x}$$ in the  domain: $$x \in [0,4]$$ Do not use for loops for the same. 

```python
def f(x:np.ndarray)->np.ndarray:
    # write your logic here 
    return (np.sin(x)**7 + np.cos(x)**5)/(np.e**x)

steps=1000 # experiment how step size matters in the graph shape
x= np.linspace(0,4, steps) # divides the whole sample space into 'steps' number of divisions uniformly
y= f(x) 
plt.plot(x,y) # label the axes, and choose appropriate plot format and style
plt.xlabel("x")
plt.ylabel('f(x)')
plt.show()
```

Question 5: Create a subplot with two plots side by side. Plot a sine wave in the first subplot. Plot a cosine wave in the
second subplot. Add labels, titles, and a legend to the plots.

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(0, 2*np.pi, 1000)  # divide the sample space [0, 2pi) in 1000 divisions
sine = np.sin(x)  # vector operations in numpy
cosine = np.cos(x)
ticks = np.arange(0, 2*np.pi+1, np.pi/2)  # provides the markings for [0, 2pi] in multiples of pi/2

plt.title('Plotting the values of sine and cosine of x wrt x', loc='center', pad=15)  # padding for good looks
plt.xticks([])
plt.yticks([])

plt.subplot(1,2,1)  # one row, two columns, first subplot
plt.plot(x, sine, 'red')
plt.legend(['sin(x)'])
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)
plt.xticks(ticks, ['0', 'π/2', 'π', '3π/2', '2π'])

plt.subplot(1,2,2)  # second subplot
plt.tick_params(
    axis='y',
    which='both',
    left=False,
    labelleft=False  # to switch off values and ticks for the right graph (looks good with the ticks off)
)
plt.plot(x, cosine, 'green')
plt.legend(['cos(x)'])
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)
plt.xticks(ticks, ['0', 'π/2', 'π', '3π/2', '2π'])

plt.show()
```
